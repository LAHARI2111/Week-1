```python
# Install essential libraries 
# !pip install pandas numpy matplotlib seaborn scikit-learn

# Importing Libraries for Data Handling & Modeling
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load Dataset
file_path = 'PB_All_2000_2021.csv'
df = pd.read_csv(file_path, sep=';')

# Basic Info
print("Shape of the dataset:", df.shape)
print("Columns in the dataset:", df.columns.to_list())
display(df.head(3))
df.info()

# Statistical Summary
display(df.describe().T.style.background_gradient(cmap='YlGnBu'))

# Check for Missing Values
null_summary = df.isnull().sum()
print("\nMissing Values:")
print(null_summary[null_summary > 0] if null_summary.sum() > 0 else "No missing values found.")

# Convert and Enrich Date Features
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y', errors='coerce')

# Sort and Add Year-Month Columns
df = (
    df.sort_values(by=['id', 'date'])
      .assign(
          year=lambda x: x['date'].dt.year,
          month=lambda x: x['date'].dt.month
      )
)

# Preview after Enrichment
print("\nPreview after Feature Enrichment:")
display(df.head(3))
print("\nUpdated Column Names:", ', '.join(df.columns))
```
